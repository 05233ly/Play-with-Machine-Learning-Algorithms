### ``matplotlib`` 基础

```python
import numpy as np 
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot(x, y)
```

```python
plt.show()
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

关于color参数：[https://matplotlib.org/2.0.2/api/colors_api.html](https://matplotlib.org/2.0.2/api/colors_api.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

关于linestyle参数：[https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html](https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15)
plt.ylim(0, 1)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!")
plt.show()
```

### Scatter Plot

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
iris.data.shape
```

```python
data = iris.data[:,:2]
```

```python
plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
plt.scatter(data[:,0], data[:,1], color="red")
plt.show()
```

```python
y = iris.target
y
```

```python
plt.scatter(data[y==0,0], data[y==0,1], color="red")
plt.scatter(data[y==1,0], data[y==1,1], color="blue")
plt.scatter(data[y==2,0], data[y==2,1], color="green")
plt.show()
```

```python
plt.scatter(data[y==0,0], data[y==0,1], color="red", marker="o")
plt.scatter(data[y==1,0], data[y==1,1], color="blue", marker="+")
plt.scatter(data[y==2,0], data[y==2,1], color="green", marker="x")
plt.show()
```

关于marker参数：[http://matplotlib.org/1.4.2/api/markers_api.html](http://matplotlib.org/1.4.2/api/markers_api.html)

```python
plt.scatter(x[::3], siny[::3])
plt.plot(x[::3], siny[::3])
plt.show()
```
